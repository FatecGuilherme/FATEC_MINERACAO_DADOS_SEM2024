```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 01 - Criação e Manipulação de Arrays

Crie um array Numpy com 10 números inteiros aleatórios entre 1 e 20 e extraia os elementos pares.

```python
dados = np.random.randint(1, 21, size=10)
pares = dados[dados % 2 == 0]
print("Array original:", dados)
print("Elementos pares:", pares)
```

# 02 - Operações Aritméticas com Arrays

Crie dois arrays de 10 números inteiros aleatórios entre 0 e 50 e realize operações de soma, subtração, multiplicação e divisão elemento a elemento.

```python
primeiroArray = np.random.randint(1, 51, size=10)
segundoArray = np.random.randint(1, 51, size=10)

print("primeiro Array:", primeiroArray)
print("segundo Array:", segundoArray)
```

```python
print("Soma:", primeiroArray + segundoArray)
print("Subtração:", primeiroArray - segundoArray)
print("Multiplicação:", primeiroArray * segundoArray)
print("Divisão:", primeiroArray / segundoArray)
```

# 03 - Indexação e Slicing

Dado um array Numpy com 20 números aleatórios entre 10 e 100, extraia os elementos entre a 5ª e a 15ª posição.

```python
dados03 = np.random.randint(10, 101, size=20)
sub_array = dados03[4:15]
print("Array original:", dados03)
print("Elementos entre a 5ª e a 15ª posição:", sub_array)
```

# 04 - Manipulação de Matrizes

Crie uma matriz 3x3 com valores aleatórios entre 0 e 1, e encontre o valor máximo e mínimo.

# 07 - Criação de DataFrame com Dados Sintéticos

Crie um DataFrame fictício de vendas contendo colunas como Produto, Quantidade (entre 1 e 20), e Preço (entre 10 e 100).

```python
produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E', 'Produto F', 'Produto G', 'Produto H', 'Produto I', 'Produto A']
quantidades = np.random.randint(1, 21, size=10)
precos = np.random.randint(10, 101, size=10)

df_vendas = pd.DataFrame({
    'Produto': produtos,
    'Quantidade': quantidades,
    'Preço': precos
})

df_vendas
```

# 08 - Filtragem de Dados

Filtre o DataFrame criado no exercício anterior para mostrar apenas as vendas onde a quantidade é maior que 10.

```python
df_filtrado = df_vendas[df_vendas['Quantidade'] > 10]
df_filtrado
```

# 09 - Operações em Colunas

Adicione uma coluna ao DataFrame com o valor total da venda (Quantidade * Preço).

```python
df_vendas['Valor Total'] = df_vendas['Quantidade'] * df_vendas['Preço']
df_vendas
```

# 10 - Agrupamento e Agregação

Agrupe o DataFrame pelo campo Produto e calcule a soma das quantidades vendidas para cada produto.

```python
df_agrupado = df_vendas.groupby('Produto')['Quantidade'].sum().reset_index()
df_agrupado
```

# 11 - Gráfico de Linhas

Crie um gráfico de linha que mostra a variação de preços de um produto fictício ao longo de 12 meses.

```python
produto = [f'Produto "X"' for i in range(12)]
meses = [f'Mês {i+1}' for i in range(12)]
precos = np.random.uniform(10, 100, size=12)

df_precos = pd.DataFrame({
    'Produto': produto,
    'Mês': meses,
    'Preço': precos
})

df_precos
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_precos['Mês'], df_precos['Preço'], marker='o', linestyle='-', color='b')
plt.title('Variação de Preços ao Longo de 12 Meses do Produto "X"')
plt.xlabel('Mês')
plt.ylabel('Preço')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

plt.show()
```

# 12 - Gráfico de Barras

Crie um gráfico de barras comparando a quantidade vendida de 5 produtos fictícios.

```python
plt.figure(figsize=(10, 6))
plt.bar(df_vendas['Produto'], df_vendas['Quantidade'], color='skyblue')
plt.title('Quantidade Vendida de Produtos Fictícios')
plt.xlabel('Produto')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.show()
```

# 13 - Histograma

Crie um histograma para mostrar a distribuição de idades em um conjunto de dados fictício de 100 pessoas.

```python
idades = np.random.randint(20, 60, size=100)

plt.figure(figsize=(10, 6))
plt.hist(idades, bins=10, edgecolor='black', color='skyblue')
plt.title('Distribuição de Idades de 100 Pessoas')
plt.xlabel('Idade')
plt.ylabel('Número de Pessoas')
plt.grid(axis='y')

plt.show()
```

# 14 - Gráfico de Dispersão

Crie um gráfico de dispersão que mostre a relação entre altura e peso em um conjunto de dados fictício.

```python
alturas = np.random.uniform(150, 190, size=100)
pesos = np.random.uniform(50, 100, size=100)

plt.figure(figsize=(10, 6))
plt.scatter(alturas, pesos, color='blue', alpha=0.7)
plt.title('Relação entre Altura e Peso')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.grid(True)

plt.show()
```

# 15 - Customização de Gráficos

# 16 - Gráfico de Barras com Seaborn

Use Seaborn para criar um gráfico de barras com os mesmos dados fictícios de quantidade vendida por produto.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Produto', y='Quantidade', data=df_vendas, color='skyblue')
plt.title('Quantidade Vendida de Produtos Fictícios')
plt.xlabel('Produto')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.show()
```

# 17 - Gráfico de Dispersão com Seaborn

Recrie o gráfico de dispersão da relação entre altura e peso usando Seaborn, incluindo uma linha de regressão.

```python
df = pd.DataFrame({
    'Altura': alturas,
    'Peso': pesos
})

plt.figure(figsize=(10, 6))
sns.regplot(x='Altura', y='Peso', data=df, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Relação entre Altura e Peso com Linha de Regressão')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.grid(True)

plt.show()
```

# 18 - Boxplot

Crie um boxplot para visualizar a distribuição das idades em um conjunto de dados fictício de 50 pessoas.

```python
idades = np.random.randint(20, 60, size=50)

df = pd.DataFrame({
    'Idade': idades
})

plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Idade'], color='lightblue')
plt.title('Distribuição das Idades de 50 Pessoas')
plt.xlabel('Idade')
plt.grid(True)

plt.show()
```

# 19 - Heatmap

Crie um heatmap para mostrar a correlação entre diferentes variáveis em um DataFrame fictício.

```python
np.random.seed(0)
data = {
    'Variável A': np.random.rand(100),
    'Variável B': np.random.rand(100),
    'Variável C': np.random.rand(100),
    'Variável D': np.random.rand(100)
}
df = pd.DataFrame(data)

correlacao = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Heatmap da Correlação entre Variáveis')
plt.show()
```

# 20 - Pairplot

Crie um pairplot para explorar relações entre várias variáveis em um conjunto de dados fictício.

```python
np.random.seed(0)
data = {
    'Variável A': np.random.rand(100),
    'Variável B': np.random.rand(100),
    'Variável C': np.random.rand(100),
    'Variável D': np.random.rand(100)
}
df = pd.DataFrame(data)

sns.pairplot(df)
plt.suptitle('Pairplot das Variáveis', y=1.02)
plt.show()
```
